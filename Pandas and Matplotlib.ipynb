```python
import pandas as pd

# Import Semua Data
df_food = pd.read_csv('Food_Supply_Quantity_kg_Data.csv')
df_protein = pd.read_csv('Protein_Supply_Quantity_Data.csv')
df_fat = pd.read_csv('Fat_Supply_Quantity_Data.csv')
df_category = pd.read_csv('Supply_Food_Data_Descriptions.csv')
```

```python
#Supply Columns
supply_drop_country = df_food.drop(df_food.columns[[0]],axis=1)
supply_columns = list(supply_drop_country.drop(df_food.loc[:,'Obesity':'Unit (all except Population)'],axis=1))
print('Supply Columns:', supply_columns)

# All Columns
all_columns = list(df_food.columns)
print('\nAll Columns:', all_columns)

#Non Supply Columns
non_supply_columns = list(df_food.drop(df_food.loc[:,'Alcoholic Beverages':'Vegetal Products'], axis=1))
print('\nNon Supply Columns:', non_supply_columns)
```

```python
# Food Columns

food = df_food.copy()
food_columns = list(food)
food_add = food_columns[1:24]
food_left = food_columns[24:]
iteration = len(food_add)

for i in range(iteration):
    food_add[i] = food_add[i] + '_food'

food_labels = ['Country'] + food_add + food_left
food.columns = food_labels


# Fat Columns
fat_drop_country = df_fat.drop(df_fat.columns[[0]],axis=1)
fat_columns = fat_drop_country.drop(fat_drop_country.loc[:,'Obesity':'Unit (all except Population)'],axis=1)

fat_list = list(fat_columns)
fat_length = len(fat_list)

for i in range(fat_length):
    fat_list[i] = fat_list[i] + '_fat'
    
fat_columns.columns = fat_list
protein_drop_country = df_protein.drop(df_protein.columns[[0]],axis=1)
protein_columns = protein_drop_country.drop(protein_drop_country.loc[:,'Obesity':'Unit (all except Population)'],axis=1)

# Protein Columns
protein_list = list(protein_columns)
protein_length = len(protein_list)

for i in range(protein_length):
    protein_list[i] = protein_list[i] + '_protein'
     
protein_columns.columns = protein_list
kolom = [food,fat_columns,protein_columns]

merging = pd.concat(kolom)
```

```python
# Sorting Obesity Terbesar ke Terkecil

obesity_rank = merging.sort_values(by='Obesity',ascending=False,ignore_index=True)
print(obesity_rank[['Country','Obesity']])

indo_rank = obesity_rank[obesity_rank['Country'] == 'Indonesia']['Obesity']
print(indo_rank)

# Indonesia memiliki persentase penduduk obesitas 6.9%
# Indonesia memiliki peringkat ke-139 (138 apabila peringkat dimulai dari 0)

print('Persentase Penduduk Indonesia Obesitas: 6.9%')
print('Peringkat Penduduk Indonesia Obesitas: 139')
```

```python
df_food['Confirmed Category'] = pd.qcut(df_food['Confirmed'], 5,
        labels=['Sangat Tidak Banyak','Tidak Banyak','Sedang','Cukup Banyak','Sangat Banyak'])

print(df_food['Confirmed Category'])

df_covid = df_food.groupby('Confirmed Category')[['Obesity','Meat','Fruits - Excluding Wine','Fish, Seafood',
                 'Animal Products','Vegetal Products']].mean()

df_covid
```

```python
# Melakukan normalisasi

def normalisasi (x):
    x_new = (x - x.min())/(x.max() - x.min())
    return x_new

df_covid_norm = df_covid.agg(normalisasi)
print(df_covid_norm)

import matplotlib.pyplot as plt

# pendefinisian y
y1 = df_covid_norm['Obesity']
y2 = df_covid_norm['Meat']
y3 = df_covid_norm['Fruits - Excluding Wine']
y4 = df_covid_norm['Fish, Seafood']
y5 = df_covid_norm['Animal Products']
y6 = df_covid_norm['Vegetal Products']

# pendefinisian x
x = np.arange(0,5)

fig,ax = plt.subplots(figsize=(12,8))
ax.bar(x,y1,width=0.1,label='Obesity')
ax.bar(x+0.1,y2,width=0.1,label='Meat')
ax.bar(x+0.2,y3,width=0.1,label='Fruits - Excluding Wine')
ax.bar(x+0.3,y4,width=0.1,label='Fish, Seafood')
ax.bar(x+0.4,y5,width=0.1,label='Animal Products')
ax.bar(x+0.5,y6,width=0.1,label='Vegetal Products')

label_kolom = [0] + list(df_covid_norm.columns)

ax.set_xticklabels(label_kolom,fontsize=10)
ax.set_title('Perbandingan Konsumsi Makanan Berdasarkan Tingkat Kasus COVID-19', 
            fontsize=15)

plt.legend(loc='upper right')
plt.style.use('default')
plt.show()
```
